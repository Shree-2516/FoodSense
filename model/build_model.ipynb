TOP 6 RESTAURANTS

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/zomato.csv')

plt.figure(figsize=(10,7))

top_rest = df['name'].value_counts().head(6)

sns.barplot(x=top_rest.index, y=top_rest.values)

plt.title("Top 6 Restaurants")
plt.xticks(rotation=30)

plt.savefig('../visuals/top_restaurants.png')
plt.show()
```

RATING DISTRIBUTION

```python
plt.figure(figsize=(10,5))

# Clean rating
df['rate'] = df['rate'].astype(str).str.replace('/5','')
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')

sns.histplot(df['rate'].dropna(), bins=20)

plt.axvline(df['rate'].mean(), color='red', label='Mean')
plt.legend()

plt.title("Rating Distribution")

plt.savefig('../visuals/rating_distribution.png')
plt.show()
```

TOP 10 RATED RESTAURANTS

```python
df_clean = df.dropna(subset=['name', 'rate'])

top_rated = df_clean.sort_values(by='rate', ascending=False).drop_duplicates('name').head(10)

plt.figure(figsize=(10,7))

sns.barplot(x=top_rated['rate'], y=top_rated['name'])

plt.title("Top Rated Restaurants")

plt.savefig('../visuals/top_rated.png')
plt.show()
```

CUISINE WORD FREQUENCY

```python
from collections import Counter

# Split cuisines
cuisine_words = df['cuisines'].dropna().str.lower().str.split(', ')

# Flatten list
all_words = [word for sublist in cuisine_words for word in sublist]

counter = Counter(all_words)

common = counter.most_common(10)

words = [i[0] for i in common]
counts = [i[1] for i in common]

plt.figure(figsize=(10,6))
sns.barplot(x=counts, y=words)

plt.title("Top Cuisines")

plt.savefig('../visuals/cuisine_freq.png')
plt.show()
```
